```python
import random as rd
import numpy as np
from random import randrange
import matplotlib.pyplot as plt
```

```python
g=[]
xlabel=[]
i=100
while 600>=i:
  xlabel.append(i)
  i+=1
var= np.random.choice([100, 200, 300, 400, 500, 600], p = [2/9, 0.25, 1/9, 0.25, 1/9, 1/18])
p=100
print (var)
while 600>=p:
  ganan=1000*var-750*p
  g.append(ganan)
  #print(ganan)
  p+=1

plt.plot(xlabel,g)
plt.title(var)
plt.show()
```

```python
promg=[]
promv=[]
g=[]
v=[]
n=1000
nn=50
i=0
ii=0
while nn>=ii:
  while n>=i:
    var= np.random.choice([100, 200, 300, 400, 500, 600], p = [2/9, 0.25, 1/9, 0.25, 1/9, 1/18])
    v.append(var)
    p=100
    while 600>=p:
      ganan=1000*var-750*p
      g.append(ganan)
      p+=1
    i+=1
    pg=sum(g)/len(g)
    pv=sum(v)/len(v)
    promg.append(pg)
    promv.append(pv)
  ii+=1


plt.plot(promv, color='orange')
plt.xlabel('Iteraciones')
plt.ylabel('Probabilidad')
plt.title('articulos comprados')
plt.show()

plt.plot(promg, color='purple')
plt.xlabel('Iteraciones')
plt.ylabel('Probabilidad')
plt.title('ganancias')
plt.show()
```
