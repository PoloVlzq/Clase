```python
import random
from random import randrange
import matplotlib.pyplot as plt
N=[]
prom=[]
nn=10000
n=0
R=500
i=0
while R>i:
  while n<nn:
    cas=0
    d=0
    tds=[]
    while cas<20:
        d=random.randrange(1,7)
        cas=cas+d
        if cas==3:
            cas=11
        if cas==13:
            cas=4
        if cas==15:
            cas=19
        if cas==17:
            cas=10
        tds.append(cas)
    N.append(len(tds))
    n=n+1
    prom.append(sum(N)/len(N))
  i+=1

plt.plot(prom, color='green')
plt.xlabel('Iteraciones')
plt.ylabel('Probabilidad')
plt.title('Convergencia de la probabilidad estimada')
plt.show()

```
