```python
import random as rd
from random import randrange
import matplotlib.pyplot as plt

n=10000
nn=50
i=0
ii=0
calls=[]
prom=[]

while nn>ii:
  while n>i:
    p=rd.randrange(1,4)
    if p==1:
      va=rd.randrange(16,25)
      calls.append(va/10)
    if p==2:
      va=rd.randrange(18,25)
      calls.append(va/10)
    if p==3:
      va=rd.randrange(14,25)
      calls.append(va/10)
    i+=1
    prom.append(sum(calls)/len(calls))
  ii+=1



plt.plot(prom, color='blue')
plt.xlabel('Iteraciones')
plt.ylabel('Probabilidad')
plt.title('Convergencia de la probabilidad estimada')
plt.show()
```
