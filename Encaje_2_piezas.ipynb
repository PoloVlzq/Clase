```python

import random
import matplotlib.pyplot as plt

n=10000
nn=500
i=0
ii=0
c=0
p=[]
prom=[]
while nn>ii:
    while n>i:
        x=random.gauss(1.5,0.0016)
        y=random.gauss(1.48,0.009)

        if x<=y:
            c=c+1
        i=i+1

    pro= (c/n)*100
    #print ("probabilidad de", pro,"%")
    p.append(pro)
    i=0
    c=0
    ii=ii+1
    prom.append(sum(p)/len(p))
#print (prom)
plt.plot(prom, color='red')
plt.xlabel('Iteraciones')
plt.ylabel('Probabilidad')
plt.title('Convergencia de la probabilidad estimada')
plt.show()

```
